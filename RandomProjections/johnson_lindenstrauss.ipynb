Show the trade-off between the minimum number of dimensions and distortion for the MNIST dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.random_projection import johnson_lindenstrauss_min_dim

# Plott for the Johnson-Lindenstrauss lemma:
size_of_dataset = 50000
x = np.arange(0.1,1,0.1)

y = johnson_lindenstrauss_min_dim(n_samples=50000, eps=x)

plt.plot(x,y,label='n_samples = 50000')
plt.xlabel("Distortion")
plt.ylabel("Minimum number of dimensions")
plt.legend()
plt.savefig('tradeoff between dimension and distortion.png')

plt.show()
```
